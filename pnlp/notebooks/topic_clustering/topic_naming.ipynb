```python
# sort csv into clusters, remove unnecessary values
keyword_list = pd.read_csv("./data_tm.csv", usecols= ["Question","Cluster ID","Cluster Keywords"])
keyword_list = keyword_list.drop_duplicates()
keyword_list = keyword_list.T

# sort data into question 1 and question 2, get the keywords for each
indices_q1 = [c for c in keyword_list if keyword_list[c][0] == "Please tell us what is working well."]
keywords_q1 = keyword_list[indices_q1]
keywords_q1 = keywords_q1.drop(index = "Question")
keywords_q1.columns = keywords_q1.iloc[0]
keywords_q1 = keywords_q1.drop(index = "Cluster ID")

indices_q2 = [c for c in keyword_list if keyword_list[c][0] == "Please tell us what needs to be improved."]
keywords_q2 = keyword_list[indices_q2]
keywords_q2 = keywords_q2.drop(index = "Question")
keywords_q2.columns = keywords_q2.iloc[0]
keywords_q2 = keywords_q2.drop(index = "Cluster ID")
```

```python
# function to find most similar word (title) for a list of keywords
def find_most_similar_cosmul(keywords):
    most_similar = {}
    for cluster in keywords.columns:
        keywords[cluster].dropna(inplace = True)
        most_similar[cluster] = en_model.most_similar_cosmul(positive=keywords[cluster][0], 
                                                                         negative=None, topn=3)
    return most_similar
```

```python
def most_similar_w_negatives(keywords):
    most_similar = {}
    for cluster in keywords.columns:
        keywords[cluster].dropna(inplace = True) 
        negative_words = set([])
        for other in keywords.columns:
            if cluster != other:
                for w in keywords[other][0]:
                    negative_words.add(w)
        most_similar[cluster] = en_model.most_similar_cosmul(positive= keywords_q1_cleaned[cluster][0], 
                                                                     negative= list(negative_words), topn=3)
    return most_similar
```

```python
from pprint import pprint
# calculate 3 words as title for keywords of the clusters for question 1 and 2 

titles_q1 = find_most_similar_cosmul(keywords_q1_cleaned)
titles_q2 = find_most_similar_cosmul(keywords_q2_cleaned)
pprint(titles_q1)
pprint(titles_q2)

# calculate 3 words as title for keywords taking all other keywords as negatives

titles_q1_n = most_similar_w_negatives(keywords_q1_cleaned)
titles_q2_n = most_similar_w_negatives(keywords_q2_cleaned)
pprint(titles_q1_n)
pprint(titles_q2_n)
```
